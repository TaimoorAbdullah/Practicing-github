# Titanic Survival Prediction - Logistic Regression

This notebook analyzes the Titanic dataset and prepares it for logistic regression modeling.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the Titanic training dataset
data = pd.read_csv('titanic_train.csv')
data.head()
```

```python
# Check for missing values in the dataset
data.isnull().any()
```

```python
# Fill missing Age values with the mean age
data['Age'].fillna(data['Age'].mean(), inplace=True)
```

```python
# Verify that Age column no longer has missing values
data.isnull().any()
```

```python
# Check the Embarked port distribution (remove NaN)
data['Embarked'].dropna()
```

```python
# Check survival distribution
data['Survived'].value_counts()
```

```python
# Drop columns that are not useful for modeling
data.drop(['PassengerId', 'Ticket', 'Name'], axis=1, inplace=True)
```
